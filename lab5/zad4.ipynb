```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
(train_set, test_set) = train_test_split(df.values, train_size=0.7, random_state=10)
train_input = list(map(lambda x: x[:-1], train_set))
train_label = list(map(lambda x: x[4], train_set))

test_input = list(map(lambda x: x[:-1], test_set))
test_label = list(map(lambda x: x[4], test_set))
```

```python
classifiers = [
    tree.DecisionTreeClassifier(),
    KNeighborsClassifier(n_neighbors=3),
    KNeighborsClassifier(n_neighbors=5),
    KNeighborsClassifier(n_neighbors=11),
    GaussianNB()
  ]
```

```python
def calculate(classifier):
  clf = classifier.fit(train_input, train_label)
  pred = clf.predict(test_input)
  counter = 0
  for i, j in zip(test_label, list(pred)):
      if i == j:
          counter += 1
  print(counter, counter/len(pred))
  print(confusion_matrix(test_label, pred, labels=["virginica", "setosa", "versicolor"]))
```

```python
for i in classifiers:
    calculate(i)
```
