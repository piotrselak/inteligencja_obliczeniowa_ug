```python
import pandas
import matplotlib.pyplot as plt
```

```python
x = miasta["Rok"]
y = miasta["Gdansk"]
plt.plot(x, y, "-r", marker="o")
plt.title("Ludnosc w Gdansku")
plt.xlabel("Lata")
plt.ylabel("Liczba ludnosci (w tys.)")
```

```python
x = miasta["Rok"]
plt.plot(x, miasta["Gdansk"], "-r", marker="o", label="Gdansk")
plt.plot(x, miasta["Poznan"], "-b", marker="o", label="Poznan")
plt.plot(x, miasta["Szczecin"], "-g", marker="o", label="Szczecin")
plt.title("Ludnosc w miastach Polski")
plt.xlabel("Lata")
plt.ylabel("Liczba ludnosci (w tys.)")
plt.legend()
```

```python
def standarize(frame):
    mean = frame.mean()
    std = frame.std()
    return frame.map(lambda x: (x-mean)/std)

for i in ["Gdansk", "Poznan", "Szczecin"]:
   stand = standarize(miasta[i])
   print(i, stand.mean(), stand.std()) 
```

```python
def normalize(frame):
    minimum = frame.min()
    maximum = frame.max()
    return frame.map(lambda x: (x-minimum)/(maximum - minimum))

for i in ["Gdansk", "Poznan", "Szczecin"]:
   stand = normalize(miasta[i])
   print(i, stand.min(), stand.max()) 
```
